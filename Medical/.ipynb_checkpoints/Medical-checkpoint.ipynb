```python
import pandas as pd
import numpy as np
Medical_compare = pd.read_csv('compareFormula4.csv',encoding = "gbk")  # 编码格式非utf-8(国际通用码)，是中文gbk模式
Medical_compare.head()
# print(type(Medical_compare))

LeftId = Medical_compare['LeftFormulaID']
RightId = Medical_compare['RightFormulaID']
print(type(LeftId))
# print(LeftId)
# print(RightId)

# 想尝试用差值的方法来比较LeftId和RightId是否相同，如为0则相同
Medical_Id_difference = LeftId - RightId
Medical_compare['Medical_Id_difference'] = Medical_Id_difference

# print(Medical_compare['Medical_Id_difference'])

print('------排序-----')
Medical_compare.sort_values("Medical_Id_difference",inplace=True)
# print(Medical_compare['Medical_Id_difference'])

print('-----Data_Frame转换为List----')
# print(LeftId.extend(RightId))
LeftId = np.array(LeftId).tolist() # 把Data_Fram 转换为list
RightId = np.array(RightId).tolist()
# Calling DataFrame constructor on list 
# df = pd.DataFrame(lst) 把list转换为DataFeame
print(type(LeftId))
print(type(RightId))
print('-----原始LeftId和RightId的数量-----')
print(len(LeftId))
print(len(RightId))

print('-----未去重的LeftId与RightId的差集')
ret = [ i for i in LeftId if i not in RightId ]  # List差集运算
# print(ret)
print(len(ret))

print('------RightId和LeftId自身去重')
LeftId = list(set(LeftId)) 
Right = list(set(RightId))
print(len(LeftId))
print(len(RightId))

print('-----去重后的LeftId与RightId的差集')
ret1 = [ i for i in LeftId if i not in RightId ] # List差集运算
ret1.sort()  # 排序
# print(ret1)
print(len(ret1))
# print(type(ret1))

# 找出对应LeftId的Name/未去重的
n = 0
for m in ret1:
    print(m)
    print(Medical_compare.LeftFormulaName[Medical_compare.LeftFormulaID == m])
    n = n + 1

print(n)
```

```python
#在python3对列表的处理中，会经常使用到Python求两个list的差集、交集与并集的方法。
#下面就以实例形式对此加以分析。
#一.两个list差集
#如有下面两个数组：
a = [1,2,3]
b = [2,3,5,6]

#想要的结果是[1]

#下面记录一下三种实现方式：

#1. 正常的方式
ret = []
for i in a:
    if i not in b:
        ret.append(i)

#2.简化版

ret = [ i for i in a if i not in b ]

#3.高级版

ret = list(set(a) ^ set(b))

#4.最终版

print (list(set(b).difference(set(a)))) # b中有而a中没有的

 
#二.两个list并集

 
print (list(set(a).union(set(b))))

# 三.两个list交集
print (list(set(a).intersection(set(b))))
```
